```python
from sklearn.manifold import MDS
from sklearn.metrics import euclidean_distances

import pandas
```

```python
data = pandas.read_csv("c_0000.csv").drop(["vx", "vy", "vz", "m", "id"], axis=1)[:1000].apply(pandas.to_numeric)
data -= data.mean()
```

```python
mds = MDS(n_components=3, max_iter=300000, eps=1e-9).fit(data)
pos = mds.embedding_
stress = mds.stress_
```
